```python
import nltk
```

```python
nltk.download('cess_esp')
```

```python
#separar las palabras por espacios
texto_corpus = ' '.join(nltk.corpus.cess_esp.words())
```

```python
#Tokenizar
tokens = nltk.word_tokenize(texto_corpus)
```

```python
#guardar en una variable las palabras en minuscula y sin tildes
tokens = [palabra.lower()for palabra in tokens if palabra.isalnum()]
```

```python
#realizar la frecuencia de los datos recolectados
frecuencia = nltk.FreqDist(tokens)
print("Palabras más frecuentes:")
for palabra, freq in frecuencia.most_common(20):
    print(f"{palabra}: {freq}")
```

```python
import matplotlib.pyplot as plt

# Obtener las etiquetas (palabras) y los valores (frecuencias)
etiquetas = [palabra for palabra, freq in frecuencia.most_common(20)]
valores = [freq for palabra, freq in frecuencia.most_common(20)]
# Gra+àfico de los resultados del análisis de frecuencia
plt.plot(etiquetas, valores, marker='o', linestyle='-', color='b')
plt.title("Palabras más frecuentes en el corpus cess_esp")
plt.xlabel("Palabra")
plt.ylabel("Frecuencia")
plt.xticks(rotation = 45)
plt.show()
```

```python
#definicion de variables con sentimientos
palabras_positivas = ['bueno', 'feliz', 'alegre']
palabras_negativas = ['malo', 'triste', 'enfado']

#inciación de contador para las palabras
sentimientos = {'positivo': 0, 'negativo': 0}
for palabra in tokens:
  if palabra.lower() in palabras_positivas:
      sentimientos['positivo'] += 1
  elif palabra.lower() in palabras_negativas:
      sentimientos['negativo'] += 1
```

```python
#grafica de sentimientos
plt.figure(figsize=(6, 4))
plt.bar(sentimientos.keys(), sentimientos.values(), color=['green', 'red'])
plt.title('Análisis de sentimientos en el corpus cess_esp')
plt.xlabel('Sentimiento')
plt.ylabel('Cantidad de palabras')
plt.show()
```

```python
#definir el diccionario- construyendo un archivo json
temas = {'deportes': 0, 'política': 0, 'entretenimiento': 0}
#contador
for palabra in tokens:
    if palabra.lower() in ['fútbol', 'baloncesto', 'tenis']:
        temas['deportes'] += 1
    elif palabra.lower() in ['política', 'gobierno', 'elecciones']:
        temas['política'] += 1
    elif palabra.lower() in ['cine', 'música', 'televisión']:
        temas['entretenimiento'] += 1
```

```python
plt.figure(figsize=(6, 4))
plt.bar(temas.keys(), temas.values(), color=['blue', 'orange', 'purple'])
plt.title('Análisis de temas en el corpus cess_esp')
plt.xlabel('Tema')
plt.ylabel('Cantidad de palabras')
plt.show()
```
